```python
#lets form the document that contains the words and the categories associated with them
document=[]
for category in movie_reviews.categories():
  for fileid in movie_reviews.fileids(category):
    document.append((movie_reviews.words(fileid), category))
document[0:10]
```

```python
#lets shuffle the studd int the document  
import random
random.shuffle(document)
document[0:5]
#now all we have to do is to clean the data
#means we will remove all the stop words and then we have to lemmatize
```

```python
#lets import the stop words
from nltk.corpus import stopwords
stop=stopwords.words("english")
import string
punc=string.punctuation
punc=list(punc)
stop=stop+punc
len(stop)
#so we have collected all the stop words and the punctuations
```

```python
#lets import the lemmatize function
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
#lets import the pos_tag function
from nltk import pos_tag
from nltk.corpus import wordnet
```

```python
def get_simple_pos(tag):
  if tag.startswith('J'):
    return wordnet.ADJ
  elif tag.startswith('V'):
    return wordnet.VERB 
  elif tag.startswith('N'):
    return wordnet.NOUN
  elif tag.startswith('R'):
    return wordnet.ADV
  else:
    return wordnet.NOUN
```

# be very careful handling the pos_tag as it takes the input as list not as string

```python
def clean_review(words):
  #the words is in the form of list
  output_words=[]
  for w in words:
    if w.lower() not in stop:
      pos=pos_tag([w]) #we need to conert this in the form of list.
      clean_word=lemmatizer.lemmatize(w, pos=get_simple_pos(pos[0][1])) #the pos take input in the form of array, and the 1st index of the tuple is the pos_tag of that word
      output_words.append(clean_word)
  return output_words
```

```python
document_clean=[]
 #lets get all the clean words in the document clean
for w in document:
  words_in_w=w[0]
  category_in_w=w[1]
  #we have to transform the category_in_w to list of clean word
  clean_words=clean_review(words_in_w)
  document_clean.append((clean_words, category_in_w))
document_clean[0]   
```

```python
document_clean[2][1]
```

```python
#so we already have out clean document and we dont need to shuffle any further, so all we can do is assign the first 1500 datapoints to training and 500 to testing
training_document=document_clean[0:1500]
testing_document=document_clean[1500:]
```

```python
#so we have to train our model using the traing datasets only
all_words=[]
for doc in training_document:
  all_words+=doc[0]
len(all_words)
```

```python
from nltk import FreqDist
```

```python
freq=nltk.FreqDist(all_words)
freq
```

```python
#lets get the most common 3000 words
common=freq.most_common(3000)
#and these words will form the feature for the naive bayes algorithm
features=[i[0] for i in common]
len(features)
```

```python
def get_dictionary(datapoints, features):
  ans=[]
  for point in datapoints:
    words=point[0]
    category=point[1]
    feature_presence={} #this must be sent in the form of dictionary
    for feature in features:
      if feature in words:
        feature_presence[feature]=True
      else:
        feature_presence[feature]=False
    ans.append((feature_presence, category))
  return ans
```

```python
#now we have the 3000 top words which also act as the features, all we can do is just create the dictionary for each document point and just 
#fund whether that feature/ word is present or not
training_dictionary=get_dictionary(training_document,features)
```

```python
testing_dictionary=get_dictionary(testing_document,features)
```

```python
#now the data is ready with us, all we have to do is implement the naive bayes algorithm which is inbuild specifically for this scenario
from nltk import NaiveBayesClassifier
classifier=NaiveBayesClassifier.train(training_dictionary)
```

```python
nltk.classify.accuracy(classifier, testing_dictionary)
```
