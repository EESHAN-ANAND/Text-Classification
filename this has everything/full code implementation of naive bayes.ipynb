```python
from sklearn import datasets
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#the program to find the dictionary if the data are given in DataFrame format
def solution(x_train, y_train):
  result={}
  result["total data"]=len(y_train)
  all_class=set(y_train)
  for perticular_class in all_class:
    result[perticular_class]={}
    #for all the class i need to get the rows corresponding to each class
    #single class here is defined by perticular class
    row_index_of_perticular_class=y_train[y_train==perticular_class].index
    result[perticular_class]["total count"]=len(row_index_of_perticular_class)
    x_train_class=x_train.iloc[row_index_of_perticular_class]
    y_train_class=y_train.iloc[row_index_of_perticular_class]
    # now lets traverse through each feature in the x_train feature
    for j in range(x_train_class.shape[1]):
      result[perticular_class][j]={}
      for k in range(y_train_class.shape[0]):
        if x_train_class.iat[k,j] in result[perticular_class][j]:
          result[perticular_class][j][x_train_class.iat[k,j]]+=1
        else:
          result[perticular_class][j][x_train_class.iat[k,j]]=1

  return result
```

```python
def probability(dictionary, x, perticular_class):
  total_count=dictionary[perticular_class]['total count']
  total_data=dictionary["total data"]
  probability=np.log(total_count)-np.log(total_data)
  features=list(dictionary[perticular_class].keys())
  features.remove("total count")
  c=0
  for feature in features:
    x_entity=x[c]
    # print(x_entity)
    #now i am adding one so as to handle the border cases as well
    count_with_perticular_value=0
    if (x_entity in dictionary[perticular_class][feature]):
      count_with_perticular_value=dictionary[perticular_class][feature][x_entity]+1
    else:
      count_with_perticular_value=1
    count_of_the_feature=dictionary[perticular_class]["total count"]+len(dictionary[perticular_class][feature].keys())
    probability+=np.log(count_with_perticular_value)-np.log(count_of_the_feature)
    c+=1

  return probability
```

```python
def predictSinglePoint(dictionary, x):
  all_class=list(dictionary.keys())
  all_class.remove("total data")
  p=-1000
  best_class=-1
  for perticular_class in all_class:
      probability_of_this_class=probability(dictionary, x ,perticular_class)
      if (probability_of_this_class>p):
        best_class=perticular_class
        p=probability_of_this_class
  return best_class
```

```python
#lets say we want to predict the result for the x_test
#lets say we have x_test in the form of numpy array
def predict(dictionary, x_test):
  y_pred=[]
  for x in x_test:
    x_class=predictSinglePoint(dictionary, x)
    y_pred.append(x_class)
  return y_pred
```

```python
def makeLabelled(column):
    second_limit = column.mean()
    first_limit = 0.5 * second_limit
    third_limit = 1.5*second_limit
    for i in range (0,len(column)):
        if (column[i] < first_limit):
            column[i] = 0
        elif (column[i] < second_limit):
            column[i] = 1
        elif(column[i] < third_limit):
            column[i] = 2
        else:
            column[i] = 3
    return column
```
