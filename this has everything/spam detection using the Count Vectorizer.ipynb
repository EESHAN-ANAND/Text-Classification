# Well we will use the inbuild function so as to transform our description in the form of matrix.
# Each row of a matrix represents count of word in that line(description). With the help of the count of the words we can actually the algorithm of naive bayes 

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/naive bayes/this has everything/spam.csv")
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(data["Message"], data["Category"], test_size=0.25, random_state=2)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(x_train)
x_train_count.toarray()[:3], x_train_count.shape
```

```python
x_test_count=v.transform(x_test)
x_test_count.toarray(), x_test_count.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
algo=MultinomialNB()
algo.fit(x_train_count, y_train)
algo.score(x_train_count, y_train), algo.score(x_test_count, y_test) 
```
